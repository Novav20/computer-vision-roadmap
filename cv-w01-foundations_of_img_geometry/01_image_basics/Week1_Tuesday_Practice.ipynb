# Task 1: Visualization
1. Setup Notebook: Create Week1_Tuesday_Practice.ipynb. Import cv2, numpy, and matplotlib.pyplot. Matplotlib is often preferred for inline plotting in notebooks.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ensure images load correctly in matplotlib (OpenCV loads BGR instead of RGB)
def load_and_convert(image_path):
    img = cv2.imread(image_path)
    if img is None:
        return None
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Load images
img_1_rgb = load_and_convert("./civilization.png")
img_1_affine = load_and_convert("./output_images/affine_warped_output.jpg")
img_2_rgb = load_and_convert("./singing.png")
img_2_affine = load_and_convert("./output_images/perspective_warped_output.jpg")

if any([img_1_rgb is None, img_1_affine is None, img_2_rgb is None, img_2_affine is None]):
    print("Error loading images")
    exit()

# Display images side by side
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(img_1_rgb)
ax[0].set_title("Original Image")
ax[1].imshow(img_1_affine)
ax[1].set_title("Affine Warped Image")
plt.show()

fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(img_2_rgb)
ax[0].set_title("Original Image")
ax[1].imshow(img_2_affine)
ax[1].set_title("Perspective Warped Image")
plt.show()
```

# Task 2: Compose Transforms

```python
def get_augmented_affine_matrix(M_2x3):
    """
    Given a 2x3 affine matrix, returns a 3x3 augmented affine matrix.
    """
    return np.vstack(M_2x3, np.array([0, 0, 1]))

def compose_transforms_3x3 (M2, M1):
    """
    Given two 3x3 affine matrices, returns their composition (M2 applied after M1)."""
    return M2 @ M1
```
